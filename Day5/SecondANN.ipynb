```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
```

```python
data = fetch_california_housing()
```

```python
data.target
```

```python
scaler=StandardScaler()
```

```python
df = pd.DataFrame(

    data=data.data
    
)
df["target"]=data.target
```

```python
df.describe()
```

```python
df[0]=scaler.fit_transform(df.iloc[:,0:1])
df[1]=scaler.fit_transform(df.iloc[:,1:2])
df[2]=scaler.fit_transform(df.iloc[:,2:3])
df[3]=scaler.fit_transform(df.iloc[:,3:4])
df[4]=scaler.fit_transform(df.iloc[:,4:5])
df[5]=scaler.fit_transform(df.iloc[:,5:6])
df[6]=scaler.fit_transform(df.iloc[:,6:7])
df[7]=scaler.fit_transform(df.iloc[:,7:8])
```

```python
model = Sequential()
df
```

```python
model.add(
    
    Dense(units=64, input_dim=7,activation="relu")
    
)
model.add(Dropout(0.9))
```

```python
model.add(

    Dense(units=1,activation="relu")

)
```

```python
model.compile(loss='mean_squared_error',optimizer="sgd",metrics=["acc"])
```

```python
x=keras.callbacks.TensorBoard(log_dir='./logs',update_freq='epoch')
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(df.iloc[:,0:7],df.target,test_size=0.2,random_state=9)
```

```python
Xtrain.shape
```

```python
model.fit(Xtrain,Ytrain,epochs=5,batch_size=500,callbacks=[x])
```

```python
model.evaluate(Xtest,Ytest)
```
