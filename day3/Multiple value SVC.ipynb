```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.svm import SVC
from decimal import Decimal
import pickle
from sklearn.metrics import accuracy_score
```

```python
load_iris()
```

```python
iris = load_iris()
```

```python
iris.feature_names
```

```python
df= pd.DataFrame(

    data=iris.data,
    columns=iris.feature_names
)
```

```python
df["flower"]=iris.target
```

```python
df
```

```python
lr1=LinearRegression()
lr2=LinearRegression()
lr3=LinearRegression()
lr4=LinearRegression()
```

```python
c1= df["sepal length (cm)"].values.reshape(-1,1)
c2= df["sepal width (cm)"].values.reshape(-1,1)
c3= df["petal length (cm)"].values.reshape(-1,1)
c4= df["petal width (cm)"].values.reshape(-1,1)
c5= df["flower"].values.reshape(-1,1)
```

```python
lr1.fit(c1,c5)
lr2.fit(c2,c5)
lr3.fit(c3,c5)
lr4.fit(c4,c5)
```

```python
plt.plot(c1,lr1.predict(c1),color="red",label="sepal length")
plt.plot(c2,lr2.predict(c2),color="green",label="sepal length")
plt.plot(c3,lr3.predict(c3),color="blue",label="sepal length")
plt.plot(c4,lr4.predict(c4),color="yellow",label="sepal length")
```

```python
plt.scatter(c1,c5,color="red",label="sepal length")
plt.scatter(c2,c5,color="green",label="sepal width")
plt.scatter(c3,c5,color="blue",label="petal length")
plt.scatter(c4,c5,color="yellow",label="petal width")
plt.legend(loc="best")
```
